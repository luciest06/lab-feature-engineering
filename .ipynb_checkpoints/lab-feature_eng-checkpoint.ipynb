```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

**Check the shape of your data**

```python
spaceship.shape
```

**Check for data types**

```python
spaceship.dtypes
```

**Check for missing values**

```python
spaceship.isna().sum()
```

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continouos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value. 

```python
spaceship=spaceship.dropna()
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
spaceship['Cabin'] = spaceship['Cabin'].apply(lambda x: x.split('/')[0])

spaceship['Cabin'].value_counts()
```

- Drop PassengerId and Name

```python
spaceship= spaceship.drop(['PassengerId','Name'],axis=1)

spaceship.columns
spaceship.shape
```

```python
spaceship.nunique()
```

- For non-numerical columns, do dummies.

```python
spaceship.head()
```

```python
features = spaceship.drop('Transported', axis=1)
target = spaceship[['Transported']]

x_train,x_test,y_train,y_test = train_test_split(features,target,test_size=0.20,random_state=0)

from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(sparse_output=False)
ohe.fit(x_train[['HomePlanet','CryoSleep','Cabin','Destination','VIP']])
x_train_trans_np = ohe.transform(x_train[['HomePlanet','CryoSleep','Cabin','Destination','VIP']])
x_test_trans_np = ohe.transform(x_test[['HomePlanet','CryoSleep','Cabin','Destination','VIP']])

x_train_trans_df = pd.DataFrame(x_train_trans_np, columns=ohe.get_feature_names_out(), index= x_train.index)
x_test_trans_df = pd.DataFrame(x_test_trans_np, columns=ohe.get_feature_names_out(), index= x_test.index)

display(x_train_trans_df)
display(x_test_trans_df)
```

**Perform Train Test Split**

```python
x_train_final_df = pd.concat([x_train_trans_df,x_train[['RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']]],axis=1)
x_train_final_df

x_test_final_df = pd.concat([x_test_trans_df,x_test[['RoomService','FoodCourt','ShoppingMall','Spa','VRDeck']]],axis=1)
x_test_final_df       

```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(x_train_final_df,y_train)
```

- Evaluate your model's performance. Comment it

```python
knn.score(x_test_final_df,y_test)

#quite good model's performance, better than before the dummification.
```

```python
pred = knn.predict(x_test_final_df)
pred
```

```python
y_test.values
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt

confusion_matrix = metrics.confusion_matrix(y_test.values, pred)
confusion_matrix

cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [0, 1])

cm_display.plot()
plt.show()
```
